# Prerequisites

### Network & Internet
- Webserver: a computer that hosts a website
- Search engine: service to find web pages
- Network: a collection of machines that can communicate to each other
    - Network switch: route messages (like a 3rd party service) from one machine to another
        - These network switches are limited in scale, hence we have many networks!
        - But using lots of routers is just bad: limited band width, too detailed, not secured
            - Security: is loose for LAN, for between networks, can add firewalls, VPNs,...
    - Router: connects many switches
    - Switch vs. Router:
        - Switches connect devices
        - Routers connect networks
- ISP: Internet Service Provider
    - a type of company
    - manages some special routers
    - these routers are connected
    - these routers can access other routers (from other ISPs)
    - own network infrastructure: cell tower, satellite links, cell towers, etc.
- Modem: Modulator Demodulator
    - Modulate: Turn digital data into a form that rides on the carrier wave (for sending).
    - Demodulate: Extract the digital data from the modulated signal (for receiving).
- Is the Internet and the Web to same thing?
    - No.
    - Internet
        - privately owned by multiple companies
        - the infrastructure (routers, cables, satellites, cell towers)
    - Web
        - a service that runs on top of the internet
        - access and share linked documents and resources using methods
            - machines that can send messages to web browsers
        - Web servers
    - Analogy
        - Internet: the pipes.
        - Web: one kind of water flowing through them (cat videos).
        - Browser: your drinking glass.
    - Example
        - Program shared via LAN in a company  Not the web
            - Thats just files moving over the local network (maybe via SMB, FTP, or some internal app).
            - It never touches HTTP/HTTPS or the public internet.
        - Cat videos on a website  The web
            - Thats using HTTP/HTTPS to fetch media from a web server.
            - Its running on top of the internet infrastructure and using web protocols.
- Intranets and Extranets: both use internet tech
    - Intranet: private network,isolated from the public internet, used to share internal resources
    - Extranet: intranet with controlled external access (still private, certain parts are accessible)
        - Connection
            - Via cables
            - Via VPN:
                - handles encryption -> security (without, data travels in packets that can be read)
    - So the Intranet, Extranet and Internet have varying levels of publicity


### Web Browsers, DNS
- DNS (Domain Name System)
    - phonebook for the Internet
    - browser look at DNS to find the website's IP address
    - owner: some companies (ICANN, Registries, Registars)
- ICANN: decides who operates top-level domains (TLDs) like .com, .org, .xyz
- TLD Registries: manage TLDs, need contract with ICANN
- Registars: retailers (GoDaddy, Namecheap, Google Domains), need agreement from registries
    - interface between client and registry
    - will pay the registry a wholesale fee for each domain registration/renewal

- Hypertext Transfer Protocol: method for clients and servers to speak to each other
- Files: two types of files in a website
    - Code
    - Assets: not code that the browser interpret
- Data is sent in thousands of small chunks called packets:
    - Flexibility: each packet can take a different route, optimize travel (without getting stuck)
    - quicker to replace small chunks

- HTTP
    - a protocol
    - uses verbs to perform actions
- HTTP request
    - if a router is a mail sorter, an HTTP request is a mail
    - has a status code and a version (Ex: HTTP2/200)
    - headers: additional info about the response
- URL: Uniform Resource Locators
    - define locations of resources on the internet
    - a web address + protocol
    - Ex: https://developer.mozilla.org/en-US/
        - protocol: https
        - domain name: developer.mozilla.org
        - path: /en-US/
